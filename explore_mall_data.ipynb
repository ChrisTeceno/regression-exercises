```python
import explore
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
```

```python
df = explore.get_mall_data()
```

```python
df = df.replace(
    ["Female", "Male"],
    [ 1, 0]
)
# replace gender with is_female to correlate with above change
df.rename(columns={"gender": "is_female"}, inplace=True)
```

```python
x_train, y_train, x_validate, y_validate, x_test, y_test, train, test, validate = explore.split_continuous(df, 'spending_score')
```

```python

columns = ['is_female','age', 'annual_income', 'spending_score']
explore.plot_variable_pairs(train,columns)
```

```python
columns = ['spending_score', 'is_female','age', 'annual_income']

sns.jointplot(kind="reg", data=train[columns], x=columns[0], y=columns[1])
```
